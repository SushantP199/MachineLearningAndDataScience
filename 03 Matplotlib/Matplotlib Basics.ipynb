# Video 3

```python
days = [1, 2, 3, 4, 5]
```

```python
organic  = [8, 9, 8, 3, 3]
ads      = [8, 9, 7, 12, 8]
affiliate= [3, 5, 5, 4, 3]
seller   = [9, 5, 8, 9, 14]
```

```python
plt.plot([], [], color='#F97F51', label='organic')
plt.plot([], [], color='#1B9CFC', label='ads')
plt.plot([], [], color='#EAB543', label='affiliate')
plt.plot([], [], color='#55E6C1', label='seller')
plt.xlabel('No of Days')
plt.ylabel('Sales')
plt.title('Amazon Sales in every 5 days')
plt.legend()
plt.stackplot(days, organic, ads, affiliate, seller, colors=['#F97F51', '#1B9CFC', '#EAB543', '#55E6C1'])
```
