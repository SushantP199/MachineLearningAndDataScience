```python
days = [1, 2, 3, 4, 5]
```

```python
organic  = [8, 9, 8, 3, 3]
ads      = [8, 9, 7, 12, 8]
affiliate= [3, 5, 5, 4, 3]
seller   = [9, 5, 8, 9, 14]
```

```python
plt.plot([], [], color='#FEA47F', label='organic')
plt.plot([], [], color='#1B9CFC', label='ads')
plt.plot([], [], color='#3B3B98', label='affiliate')
plt.plot([], [], color='#EAB543', label='seller')
plt.legend()
plt.stackplot(days,organic, ads, affiliate, seller, colors=['#FEA47F', '#1B9CFC', '#3B3B98', '#EAB543'])
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Amazon Sales')
```
