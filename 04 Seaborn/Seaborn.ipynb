```python
import seaborn as sb
```

```python
myTitanic = sb.load_dataset('titanic')
```

```python
myTitanic.head()
```

```python
myTitanic.head(10)
```

```python
sb.distplot(myTitanic['fare'])
```

```python
sb.distplot(myTitanic['fare'], color='g', bins=10, hist=False)
```

```python
myTips = sb.load_dataset('tips')
```

```python
myTips.head()
```

```python
sb.jointplot(x='total_bill', y='tip', data=myTips, color='black', kind='reg', height=5, dropna=True)
```

```python
sb.pairplot(myTips)
```

```python
sb.pairplot(myTips, kind='reg')
```

```python
myDataFrame = sb.load_dataset('exercise')
```

```python
myDataFrame.head()
```

```python
sb.factorplot(x='diet', y='pulse',hue='time', col='kind', data=myDataFrame)
#sb.catplot()
```

```python
sb.factorplot(x='diet', y='pulse',hue='time', col='kind', data=myDataFrame, kind='strip')
#sb.catplot()
```

```python
sb.catplot(x='diet', y='pulse',hue='time', col='kind', data=myDataFrame)
```

```python
sb.catplot(x='diet', y='pulse',hue='time', col='kind', data=myDataFrame, kind='point')
```

```python
sb.catplot(x='diet', y='pulse',hue='time', col='kind', data=myDataFrame, color="red")
```
