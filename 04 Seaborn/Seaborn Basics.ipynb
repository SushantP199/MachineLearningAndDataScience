```python
import seaborn as sb
```

```python
myTitanic = sb.load_dataset('titanic')
```

```python
myTitanic.head()
```

```python
sb.distplot(myTitanic['age'])
```

```python
sb.distplot(myTitanic['age'], kde=False)
```

```python
sb.distplot(myTitanic['age'], kde=False, bins=50)
```

```python
sb.jointplot(x='pclass', y='age', data=myTitanic)
```

```python
sb.jointplot(x='pclass', y='age', data=myTitanic, kind='hex' )
```
